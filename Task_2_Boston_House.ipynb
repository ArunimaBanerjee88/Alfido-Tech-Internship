```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import joblib
```

```python
data_path = "/content/Boston_cleaned_house_prices.csv"
df = pd.read_csv(data_path)
print("Boston Housing dataset loaded successfully")

# Keep only numeric columns
df = df.select_dtypes(include=[np.number])

# Features (X) and Target (y)
X = df.drop(['price'], axis=1)
y = df['price']

# Save feature column names
feature_cols = X.columns.tolist()
joblib.dump(feature_cols, "feature_columns.pkl")
print("Feature column names saved as 'feature_columns.pkl'")

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Features standardized")

# First split: temp(80%) and test(20%)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Second split: train(60%) and validation(20%)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42)

print(f"Train set size: {len(X_train)}")
print(f"Validation set size: {len(X_val)}")
print(f"Test set size: {len(X_test)}")

# Polynomial feature transformation
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_val_poly = poly.transform(X_val)
X_test_poly = poly.transform(X_test)

# Model training
model = LinearRegression()
model.fit(X_train_poly, y_train)
print("Model training completed")

# Validation performance
val_pred = model.predict(X_val_poly)
val_mse = mean_squared_error(y_val, val_pred)
val_rmse = np.sqrt(val_mse)
val_mae = mean_absolute_error(y_val, val_pred)
val_r2 = r2_score(y_val, val_pred)

print(f"\nValidation Performance:")
print(f"  MSE  : {val_mse:.2f}")
print(f"  RMSE : {val_rmse:.2f}")
print(f"  MAE  : {val_mae:.2f}")
print(f"  R   : {val_r2:.3f}")

# Test performance
y_pred = model.predict(X_test_poly)
test_mse = mean_squared_error(y_test, y_pred)
test_rmse = np.sqrt(test_mse)
test_mae = mean_absolute_error(y_test, y_pred)
test_r2 = r2_score(y_test, y_pred)

print(f"\nTest Performance:")
print(f"  MSE  : {test_mse:.2f}")
print(f"  RMSE : {test_rmse:.2f}")
print(f"  MAE  : {test_mae:.2f}")
print(f"  R   : {test_r2:.3f}")

# Save model, scaler, and poly
joblib.dump(model, 'boston_model_poly.pkl')
print("Model saved as 'boston_model_poly.pkl'")
joblib.dump(scaler, 'scaler.pkl')
print("Scaler saved as 'scaler.pkl'")
joblib.dump(poly, 'poly_features.pkl')
print("PolynomialFeatures object saved as 'poly_features.pkl'")
```

```python
model = joblib.load('boston_model_poly.pkl')
scaler = joblib.load('scaler.pkl')
poly = joblib.load('poly_features.pkl')
def predict_house_price(input_values):
    input_scaled = scaler.transform(np.array(input_values).reshape(1, -1))
    input_poly = poly.transform(input_scaled)
    predicted_price = model.predict(input_poly)
    return np.expm1(predicted_price[0])

print("Enter the following features for prediction:")
feature_names = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']
input_values = []
for feature in feature_names:
    try:
        value = float(input(f"{feature}: "))
        input_values.append(value)
    except ValueError:
        print(f"Invalid input for {feature}. Please enter a number.")
        exit()

if len(input_values) != len(feature_names):
    print(f"Expected {len(feature_names)} features, but got {len(input_values)}. Please check your inputs.")
    exit()
predicted_price = predict_house_price(input_values)
print(f"The predicted house price is: ${predicted_price:.2f}")
```
